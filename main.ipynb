```python
import google.generativeai as genai
import os
import sys
from dotenv import load_dotenv, find_dotenv
```

```python
load_dotenv(find_dotenv(), override=True)
api_key = os.environ.get('GOOGLE_API_KEY')
genai.configure(api_key=api_key)
```

```python
from PIL import Image
file_path = input('Please enter the image file path')
if not os.path.isfile(file_path):
    print('path does not exist')
    sys.exit()

image = Image.open(file_path)
image
```

```python
model = genai.GenerativeModel('gemini-1.5-flash')
uploaded_img_file = genai.upload_file(path=file_path, display_name=file_path.split('.')[0])
chat_session = model.start_chat(history=[])

while (prompt := input('').lower()) not in ['bye', 'exit', 'quit']:
    response = chat_session.send_message([prompt, uploaded_img_file])
    print(response.text)
    print('-'*50)
```
